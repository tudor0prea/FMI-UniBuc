```python
def get_intervals(num_bins):
    bins = np.linspace(0, 256, num=5) 
    return bins
```

```python
def value_to_bins(x,bins):
    new_x= np.zeros(x.shape)
    for i,elem in enumerate(x):
        new_x[i]=np.digitsize(elem,bins)
    return new_x - 1
```

```python
bins=get_intervals(2)
x_train = values_to_bins(train_images,bins)
x_test = values_to_bins(test_images, bins)
```

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes_model=MultinomialNB();
naive_bayes_model.fit(x_train, train_labels)
naive_bayes_model.score(x_test, test_labels)
```

```python
for num_bins in [3, 5, 7, 9, 11]:
    bins = np.linspace(0, 256, num=num_bins) 
    train_images_bins = values_to_bins(train_images, bins)
    test_images_bins = values_to_bins(test_images, bins)
    naive_bayes_model = MultinomialNB()
    naive_bayes_model.fit(train_images_bins, train_labels)
    acc = naive_bayes_model.score(test_images_bins, test_labels)
    print('num bins %d has acc %f' % (num_bins, acc))
```

```python
import matplotlib.pyplot as plt
def show_miscclassified_images(model, labels, images, num=10):
    pred_labels=model.predict(images)
    count=0;
    for i in range(pred_labels.shape[0]):
        if pred_labels[i]!=labels[i]:
            image=np.reshape(images[i], (28,28))
            plt.imshow(image.astype(np.uint8), cmap='gray')
            plt.title(f'predicted {pred_labels[i]} , ground truth {labels[i]}')
            plt.show()
            count+=1
            if count == num:
                break
show_miscclassified_images(naive_bayes_model, test_labels, test_images)
```

```python
def confunsion_matrix(predicted_labels, test_labels):
    num_labels = test_labels.max() + 1
    conf_mat = np.zeros((num_labels, num_labels))
    
    for i in range(len(predicted_labels)):
        conf_mat[test_labels[i], predicted_labels[i]] += 1
    return  conf_mat
    
predicted_labels = naive_bayes_model.predict(test_images_bins)
conf_mat = confunsion_matrix(predicted_labels, test_labels)

```
